```python
import openpyxl
import random
import string
import time
import pyautogui

def write_random_data_to_excel():
    # Create a new Excel workbook and select the active sheet
    workbook = openpyxl.Workbook()
    sheet = workbook.active

    # Write random numbers and alphabets to the Excel sheet
    for row in range(1, 11):  # Writing 10 rows of data
        for column in range(1, 6):  # Writing data in 5 columns
            # Generate random numbers between 1 and 100
            random_number = random.randint(1, 100)
            # Generate random alphabets
            random_alphabet = random.choice(string.ascii_uppercase)

            # Write the random number and alphabet to the cell
            sheet.cell(row=row, column=column, value=f"{random_number}{random_alphabet}")

    # Save the Excel file
    workbook.save("random_data.xlsx")

    # Close the workbook
    workbook.close()

def move_mouse_for_10_seconds():
    for _ in range(10):
        pyautogui.move(10, 10, duration=0.5)  # Move the mouse 10 pixels right and down
        pyautogui.move(-10, -10, duration=0.5)  # Move the mouse 10 pixels left and up

if __name__ == "__main__":
    write_random_data_to_excel()
    move_mouse_for_10_seconds()
```
