```python
9. How do you make a function variable refer to the global variable?

Answer:

In most programming languages, you can make a function variable refer to a global variable by explicitly indicating that 
you want to use the global variable within the function's scope. This is often done using the global keyword
(in languages like Python) or by accessing the global variable through the global object (in languages like JavaScript).

Example:
    
global_variable = 10

def my_function():
    global global_variable
    global_variable = 20

my_function()
print(global_variable)  # This will output: 20
```

```python
13. What can you do to save a programme from crashing if it encounters an error?


To prevent a program from crashing when it encounters an error, you can implement error handling mechanisms.
Proper error handling helps your program gracefully handle unexpected situations and continue functioning,
even if certain parts encounter issues. Here are some techniques you can use:
    
    
Try-Except Blocks:
    
Use try-except blocks to catch and handle exceptions that might occur within a specific code block.
This allows you to execute alternative code paths if an error occurs. For example, in Python:  
    
try:
    # Code that might raise an exception
    result = 10 / 0
except ZeroDivisionError:
    print("Error: Division by zero")
    
    
Error Logging: Implement logging mechanisms to record errors, warnings, and other relevant information.
This helps you track down issues in production environments without crashing the program.

Validation and Input Checks:
    
Before using data or user input, validate it to ensure it meets expected formats or values. 
This prevents potential errors from propagating deeper into the program.

Default Values or Graceful Fallbacks:
    
When accessing external resources (files, databases, APIs), 
have default values or fallback mechanisms in place so that if something goes wrong, the program can 
continue using a sensible alternative.

Graceful Degradation: 
    
In web applications, if a part of the system fails, ensure that the rest of the application continues to function 
without being fully dependent on the failing component.

Separation of Concerns:
    
Organize your code in a modular way. If a specific module or component encounters an error, 
it won't necessarily crash the whole program.

Testing and Debugging: Regularly test your code and fix issues during development. Use debugging tools to identify and 
correct errors before they reach production.

User-Friendly Error Messages:
Provide meaningful error messages to users so they understand what went wrong and what they can do to mitigate the issue.

Timeouts:
When interacting with external services, set timeouts to prevent your program from waiting indefinitely if
the service is slow or unresponsive.

Backup Systems: For critical applications, consider having backup systems in place that can take over if the
primary system fails.

   
    
```

```python
14. What is the purpose of the try clause? What is the purpose of the except clause?

The try and except clauses are used in programming languages like Python to implement error handling mechanisms.
They allow you to handle exceptions (errors) that might occur during the execution of a program. 
Here's the purpose of each clause:

1. Purpose of the try Clause:
    
The try clause is used to wrap a block of code that might raise exceptions. The purpose of the try clause is to attempt
to execute the code within it. If an exception occurs while executing the code within the try block,
the program's execution doesn't stop immediately, and the control is transferred to the except block
(if there's a suitable except block to handle that specific exception type)  
 
2. Purpose of the except Clause:
 
The except clause is used to define what actions should be taken when a specific type of exception occurs
within the corresponding try block. It allows you to handle exceptions gracefully, preventing the program
from crashing. You can have multiple except blocks, each targeting a different type of exception.
If an exception matches the type specified in an except block, the code within that block is executed. 
 
Here's a simple example in Python:
 
try:
    result = 10 / 0  # This will raise a ZeroDivisionError
except ZeroDivisionError:
    print("Error: Division by zero")
 
In this example, the try block attempts to perform a division by zero, which would normally raise a ZeroDivisionError.
However, the except block catches that exception and prints an error message instead of crashing the program. 
 
```
